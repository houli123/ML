# 二、梯度下降法

```python
#1 导入库
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
# 2. 读取数据
data = pd.read_csv('datasets/example3_fea2.csv')
x = data[['x1','x2']]
y = data['y']
#3. 划分数据集
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state=8)

#  添加偏置项 (全为1的列)
x_train = np.hstack([np.ones((x_train.shape[0], 1)), x_train])
x_test = np.hstack([np.ones((x_test.shape[0], 1)), x_test])
# 4. 模型训练
np.random.seed(42)
theta = np.random.randn(3, 1)  # 生成(3,1) 的元素，元素是从标准正态分布（均值为0，标准差为1）中随机抽取的
eta = 0.1  #学习率
epsilon = 1e-6  # 收敛阈值
converged = False
iteration = 0  #收敛标志表示尚未收敛
while not converged:
    # 预测
    predictions = x_train.dot(theta)
    # 计算误差
    error = predictions - y_train.values.reshape(-1, 1)
    # 计算梯度
    gradients = (2 / x_train.shape[0]) * x_train.T.dot(error)
    # 更新参数
    prev_theta = theta.copy()
    theta = theta - eta * gradients
    # 检查收敛条件
    if np.linalg.norm(theta - prev_theta) < epsilon:
        print(f"Iteration {iteration}")
        converged = True
    iteration += 1
    # 5. 模型预测
y_pred = x_test.dot(theta)
# 6. 模型评价
mse = mean_squared_error(y_test, y_pred.flatten())
# 输出截距和斜率
print("直线方程的参数为: ", theta)
print("均方误差为：", mse)
```

# 三、随机梯度下降 

```python
#1 导入库
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
# 2. 读取数据
data = pd.read_csv('datasets/example3_fea2.csv')
x = data[['x1','x2']]
y = data['y']
#3. 划分数据集
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state=8)
#4. 训练模型
sgd_reg = SGDRegressor(max_iter=50, tol=1e-6, penalty='l2', eta0=0.1, random_state=2)
#max_iter最大迭代次数
sgd_reg.fit(x_train, y_train)
# 5.预测模型
y_pred = sgd_reg.predict(x_test)  #根据x的测试集预测y的预测值
# 6. 模型评价
mse = mean_squared_error(y_test, y_pred)
print("截距值为：",sgd_reg.intercept_,",\n斜率为：" ,sgd_reg.coef_,"，\n mse值为：",mse)
```

###### max_iter=50: 最大迭代次数。SGD通过多次遍历训练数据来优化模型参数，此参数限制遍历次数。
###### tol=1e-6: 收敛阈值。如果在一次迭代中损失下降小于此阈值，则认为算法已收敛，停止迭代。
###### penalty=None: 不使用正则化。默认情况下，SGD会添加一个正则项以防止过拟合，这里禁用以便于理解基本的SGD回归。
###### eta0=1: 初始学习率。这是步长的大小，影响参数更新的速率。
###### random_state=2: 随机数种子。为了保证代码的可重现性，固定随机数种子。

# 四、小批量梯度下降

在sklearn 中，虽然 SGDRegressor 默认是使用单样本进行更新（随机梯度下降），但是它并没有直接提供小批量梯度下降的参数设置。我们可以通过设置 partial_fit 和手动分批次输入数据，来模拟小批量梯度下降的效果。

```python
# 1. 导入库
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "last"

# 2. 读取数据
data = pd.read_csv('datasets/example3_fea2.csv')
x = data[['x1', 'x2']]
y = data['y']

# 3. 划分数据集
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state=8)

# 4. 小批量设置
batch_size=32# 每个小批量的大小
n_batches=int(np.ceil(x_train.shape[0]/batch_size))# 计算小批量的数量
# 5. 初始化SGDRegressor模型
sgd_reg=SGDRegressor(max_iter=1000,tol=1e-3,penalty='l2',eta0=0.1,random_state=8)
for epoch in range(50):
    for i in range(n_batches):
        start=i*batch_size
        end=start+batch_size
        x_batch=x_train[start:end]
        y_batch=y_train[start:end]
        # x_batch
        sgd_reg.partial_fit(x_batch,y_batch)  #小批量梯度下降的模型训练方式
        
# 6. 使用小批量梯度下降进行训练
y_pred=sgd_reg.fit(x_train,y_train)
# 7. 预测模型
y_pred=sgd_reg.predict(x_test)
# 8. 模型评价
mse=mean_squared_error(y_test,y_pred)
print("截距值为：", sgd_reg.intercept_, ",\n斜率为：", sgd_reg.coef_, "，\n mse值为：", mse)
```

# 五、使用torch来实现小批量梯度下降

```python
# 1. 导入库
# !pip install torch
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

# 2. 读取数据
data = pd.read_csv('datasets/example3_fea2.csv')
x = data[['x1', 'x2']].values  # 转换为 numpy 数组
y = data['y'].values.reshape(-1, 1)  # 转换为 numpy 数组并调整形状

# 3. 划分数据集
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state=8)

# 4. 转换为 PyTorch 张量
x_train_tensor=torch.tensor(x_train,dtype=torch.float32)
x_test_tensor=torch.tensor(x_test,dtype=torch.float32)
y_test_tensor=torch.tensor(y_test,dtype=torch.float32)
y_train_tensor=torch.tensor(y_train,dtype=torch.float32)
# x_train_tensor
# 5. 创建 DataLoader，用于小批量梯度下降
train_dataset=TensorDataset(x_train_tensor,y_train_tensor)
train_loader=DataLoader(train_dataset,batch_size=32,shuffle=True,num_workers=0)
# 6. 定义线性回归模型
class LinearRegressionModel(torch.nn.Module):
    def __init__(self,n):
        super(LinearRegressionModel, self).__init__()
        self.linear = torch.nn.Linear(n, 1)

    def forward(self, x):
        x = self.linear(x)
        return x
# 7. 初始化模型和优化器
model=LinearRegressionModel(n=2)
criterion=nn.MSELoss()# 均方误差损失函数
optimizer=optim.SGD(model.parameters(),lr=0.1,momentum=0.9)# 随机梯度下降优化器，学习率为0.01
# 8. 训练模型
for epoch in range(50):
    model.train()
    for x_batch,y_batch in train_loader:
        # 前向传播
        predictions = model(x_batch)
        # 计算损失
        loss = criterion(predictions, y_batch)
        # 反向传播
        optimizer.zero_grad()  #梯度清零
        loss.backward()  
        # 参数更新
        optimizer.step()
        
    print(f'Epoch {epoch+1}/50，Loss：{loss.item():.2f}')  #损失越来越小

# 9. 测试模型
model.eval()
with torch.no_grad(): #禁用梯度计算
    y_pred = model(x_test_tensor)
    test_loss=mean_squared_error(y_test_tensor.numpy(),y_pred.numpy())
    print(f'Test MSE ：{test_loss.item():.5f}')
# 10. 输出模型参数（截距和斜率）
print("截距值为：", model.linear.bias.data.numpy(), ",\n斜率为：", model.linear.weight.data.numpy())
```
