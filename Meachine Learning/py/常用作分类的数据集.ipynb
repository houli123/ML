# 1.鸢尾花数据集

鸢尾花（Iris）数据集是机器学习和统计学中常用的一个经典数据集，主要用于分类任务。

**数据来源**：该数据集最早由罗纳德·费舍尔（Ronald Fisher）在20世纪30年代收集，用于通过测量鸢尾花的花萼长度、花萼宽度、花瓣长度和花瓣宽度等特征来区分不同种类的鸢尾花。

**样本数量和特征**：鸢尾花数据集包含了150个数据样本，分为3类，每类50个数据。每个数据样本包含了4个特征，分别是：*花萼长度*（Sepal Length）、*花萼宽度*（Sepal Width）、*花瓣长度*（Petal Length）和*花瓣宽度*（Petal Width），这些特征的单位都是厘米（cm）。

**目标数据**： 数据集的目标数据是鸢尾花的种类，共有三个类别：山鸢尾（Iris-setosa）、变色鸢尾（Iris-versicolor）和维吉尼亚鸢尾（Iris-virginica）。

```python
from sklearn.datasets import load_iris   #导入鸢尾花数据集iris
iris = load_iris()
print(iris.keys())
x = iris.data
y = iris.target
print("x的大小为：",x.shape,"\ny的大小为：",y.shape)
print(iris.DESCR)
```

# 2. 乳腺癌威斯康星（Wisconsin）诊断数据集

load_breast_cancer 是 scikit-learn（通常简写为 sklearn）库中的一个函数，用于加载乳腺癌威斯康星（Wisconsin）诊断数据集。这个数据集是一个常用的机器学习数据集，经常用于分类任务。

以下是关于这个数据集的一些主要特点：

**数据来源**：这个数据集最初由威斯康星州医院的 Dr. William H. Wolberg 收集。

**样本数量与特征**：数据集包含了 569 个样本，每个样本有 30 个特征。这些特征描述了乳腺肿瘤的各种测量值，如肿瘤的半径、纹理、对称性等。

**目标数据**：数据集的目标数据是二分类的，代表肿瘤的良性（benign）或恶性（malignant）。其中，0 表示良性肿瘤，1 表示恶性肿瘤。

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()  
print(data.keys())
x = data.data  
y = data.target
```

# 3. Mnist数据集

MNIST数据集（Mixed National Institute of Standards and Technology database）是一个广泛使用的机器学习数据集，主要用于图像分类任务，目标是识别每张图像上写的是哪个数字。

**数据集特点**：MNIST数据集中的手写数字来自250个不同的人，其中50%是高中学生，50%来自人口普查局（the Census Bureau）的工作人员。这样的设置使得数据集更具多样性，从而更能反映真实世界中的手写数字识别问题

**来源**：MNIST数据集来源于美国国家标准与技术研究所（National Institute of Standards and Technology, NIST）。它是从NIST的两个手写数字数据集Special Database 3和Special Database 1中分别取出部分图像，并经过一些图像处理后得到的。

**样本数量和特征**：MNIST数据集包含了70,000张手写数字图像，其中60,000张用于训练，10,000张用于测试。所有图像都是28x28的灰度图像，每个像素的灰度值在0到255之间。每张图像包含一个手写数字，从0到9，且只有一个数字。

**目标数据**：每张图像都附带一个标签，表示该图像上写的是哪个数字。这些标签是数字0到9之间的整数。



```python
import numpy as np
import scipy.io as sio
mnist = sio.loadmat('dataset/mnist-original.mat')
print("数据的键名为：",mnist.keys())
print("特征数据为: \n",mnist['data'])
print("目标数据为:",mnist['label'])
x = mnist["data"].T
y = mnist["label"].flatten()
print("x的尺寸为：",x.shape)
print("y的尺寸为：",y.shape)
print("类别值有：",np.unique(y))
```

```python
# 可视化，绘制多张图的方法
import matplotlib
import matplotlib.pyplot as plt
def plot_digits(instances, images_per_row=10, **options):
    size = 28
    images_per_row = min(len(instances), images_per_row)
    images = [instance.reshape(size,size) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    images.append(np.zeros((size, size * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = matplotlib.cm.binary, **options)
    plt.axis("off")
    

plt.figure(figsize=(9,9))
example_images = np.r_[x[:12000:600], x[13000:30600:600], x[30600:60000:590]] ## np.r_按列叠加数据
#x[:12000:600]表示从数组 x 中取前12000个元素，但每隔600个取一个。这将得到一个包含20个元素（因为 12000 // 600 = 20）的子数组。
#x[13000:30600:600]表示从索引 13000 开始，直到索引 30600（但不包括索引 30600）的元素，同时每隔 600 个元素取一个。
plot_digits(example_images, images_per_row=10)
plt.show()
```
