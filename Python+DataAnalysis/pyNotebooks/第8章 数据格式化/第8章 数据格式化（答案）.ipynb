#### 基本数据类型的转换

```python
#字符串转整数
i=int('2022')
#字符串转浮点数
f=float('20.19')
#浮点数转字符串
s=str(20.19)
#字符串转日期
import datetime
a='2019/10/1'
#使用strptime()函数格式化转换
date1=datetime.datetime.strptime(a,"%Y/%m/%d")
print(date1)
#按格式输出日期
date2=date1.strftime("%Y{y}%m{m}%d{d}".format(y="年",m="月",d='日'))
print(date2)
```

#### 列表、集合和字典的转换

```python
#创建列表
ls=[90,93,85,93]
#列表转换为集合
S=set(ls)
print(S)

#键列表
k=['k1','k2','k3','k4']
#值列表
v=['v1','v2','v3']
#zip组合,当两个列表的长度不一致时，多出的元素将被忽略
z=zip(k,v)
print(type(z))
#转成字典
d=dict(z)
print(d)
#创建字典
d2={"name":'zhangsan','age':'22','gender':'male','address':'shanghai'}
#将字典的key转换成列表
lst1=list(d2)
print(lst1)
#将字典的value转换成列表
lst2=list(d2.values())
print(lst2)
```

### zip() 函数
> #### 用于将可迭代的对象作为参数，将对象中对应的元素打包成一个个元组，然后返回由这些元组组成的列表。如果各个迭代器的元素个数不一致，则返回列表长度与最短的对象相同

```python
k=['k1','k2','k3','k4']
v=['v1','v2','v3','v4']
print(list(zip(k,v)))
print(dict(zip(k,v)))
```

## 8.3 中国股票交易数据项目实战
### 8.3.1 任务分析
- 逐行读入单只股票
- 利用字典存储单只股票信息
- 利用嵌套字典存储多只股票信息
- 股票信息保存到文件

### 8.3.2 任务实现
#### 逐行读入单只股票


```python
#逐行读入单支股票
with open('data\StockData.txt','r') as fp: #打开文本文件StockData.txt
    lines=fp.readlines()                #读入到字符串列表lines
print(len(lines))
print(lines)
```

#### 利用字典存储单只股票信息


```python
dict1={"age":23,"name":"zhangsan"}
list1=[]
list1.append(dict1)
print(list1)
dict1['age']=25
print(list1)
```

```python
#利用字典存储单支股票信息
StockList=list()                                #创建列表
for S in lines:                                 #遍历lines里的字符串S
    dict1=eval(S)                               #S转为字典
    StockList.append(dict1)
    cjl=dict1['成交量'][:-2]
    if(cjl.strip()!=''):
        dict1['成交量']=float(cjl)*10000
    else:del dict1['成交量'];                   #删除'成交量'这个关键字

    cje = dict1['成交额'][:-2]
    if (cje.strip() != ''):
        dict1['成交额'] = float(cje) * 10000
    else:del dict1['成交额'];

    wb = dict1['委比'][:-1]
    print(wb)
    if (wb.strip() != '' and wb.strip() != '-'): #特别处理空字符串和单个减号
        dict1['委比'] = float(wb)/100
    else:del dict1['委比'];
print(StockList)
```

#### 利用嵌套字典存储多只股票信息


```python
#利用嵌套字典存储多支股票信息
StockDict=dict()
for dic in StockList:
    StockDict[dic['股票名称']]=dic
print(StockDict)#.keys())
```

#### 股票信息保存到文件

```python
#股票信息保存到文件
import pickle
output=open('d:\股票信息字典.pkl','wb')
pickle.dump(StockDict,output)
```

## 巩固与应用
### 8.5.1 列表转化字典

```python
import datetime

#逐行读入股票信息
with open('data\宁波精达.txt','r') as fp: #打开文本文件StockData.txt
    lines=fp.readlines()                #读入到字符串列表lines

#提取类别
keys=lines[0].split(',')
del keys[-1]
count=len(lines)
StockDict=dict()
i=1
while i<count:
    strs=lines[i].split(',')
    a=strs[0][0:10]    #后面的00:00:00 不要
    # 字符串转日期
    date1 = datetime.datetime.strptime(a, "%Y-%m-%d")
    # 按格式输出日期
    date2 = date1.strftime('%Y{y}%m{m}%d{d}').format(y='年', m='月', d='日')
    dict1 = dict()
    j=3
    while j<len(keys):
        S=strs[j]
        if keys[j]=='收盘价' or keys[j]=='成交量' or keys[j]=='总市值':
            try:
                f=float(S)/10000
                f=round(f,2)
                dict1[keys[j]] = f
            except:
                dict1[keys[j]] = 'None'
        else: dict1[keys[j]]=S
        j=j+1
    StockDict[date2]=dict1
    i=i+1

print(StockDict)
```
